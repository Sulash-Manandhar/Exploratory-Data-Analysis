```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('EDA02titanic.csv')
titanic
```

```python
import matplotlib.pyplot as plt
```

# Plot

```python
titanic.plot(subplots=True, figsize =(15,12),sharex=False, sharey = False)
plt.show() #used for displaying the ouput
```

```python
titanic.age.plot(figsize=(15,12),fontsize=18, c= 'g',linestyle=':')
plt.title('Titanic - Ages', fontsize = 18)
plt.legend(loc=0, fontsize = 15)
plt.xlabel('Passenger No', fontsize = 18)
plt.ylabel('Age', fontsize = 18)
plt.grid()
plt.show()
```

# Xlim, Ylim - limit plot

```python
titanic.age.plot(figsize=(15,12),fontsize=18, c= 'r',linestyle=':',xlim = (200, 400), ylim = (20,40))
plt.title('Titanic - Ages', fontsize = 18)
plt.legend(loc=0, fontsize = 15)
plt.xlabel('Passenger No', fontsize = 18)
plt.ylabel('Age', fontsize = 18)
plt.grid()
plt.show()
```

# Xtick , Ytick

```python
xtick = [x for x in range(0, 901, 50)]
xtick
```

```python
ytick = [y for y in range(0, 81, 5)]
ytick
```

```python
titanic.age.plot(figsize=(15,12),fontsize=18, c= 'r',linestyle=':', xticks = xtick, yticks = ytick,rot = 5)
plt.title('Titanic - Ages', fontsize = 18)
plt.legend(loc=0, fontsize = 15)
plt.xlabel('Passenger No', fontsize = 18)
plt.ylabel('Age', fontsize = 18)
plt.grid()
plt.show()
```
