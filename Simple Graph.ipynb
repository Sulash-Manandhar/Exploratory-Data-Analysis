### Largest gross production 

```python
df['Gross Production'].nlargest()
```

```python
df['Gross Production'].nlargest(10).index
```

```python
highest_gross = df.iloc[df['Gross Production'].nlargest(10).index]
highest_gross
```

```python
fig = plt.figure(figsize = (15,12))
ax = fig.add_axes([0,0,1,1])
ax.bar(highest_gross['District'],highest_gross['Gross Production'])
plt.title('Top 10 Highest Gross Production Among All')
plt.xlabel('District')
plt.ylabel('Gross Production')
plt.rcParams['font.size'] = '16'
plt.show()
```

### Largest number of Students

```python
df['Students'].nlargest(10)
```

```python
highest_student = df.iloc[df['Students'].nlargest(10).index]
highest_student
```

```python
fig = plt.figure(figsize = (15,12))
ax = fig.add_axes([0,0,1,1])
ax.bar(highest_student['District'],highest_student['Students'])
plt.title('Top 10 Highest Students Among All')
plt.xlabel('District')
plt.ylabel('Student')
plt.rcParams['font.size'] = '16'
plt.show()
```

### Smallet Gross production

```python
smallest_gross = df.iloc[df['Gross Production'].nsmallest(10).index]
smallest_gross
```

```python
fig = plt.figure(figsize = (12,6))
ax = fig.add_axes([0,0,1,1])
ax.bar(smallest_gross['District'],smallest_gross['Gross Production'])
plt.title('Top 10 Smallest Gross Production Among All')
plt.xlabel('District')
plt.ylabel('Gross Production')
plt.rcParams['font.size'] = '18'
plt.show()
```

### Smallest Student population

```python
smallest_students = df.iloc[df['Students'].nsmallest(10).index]
smallest_students
```

```python
fig = plt.figure(figsize = (12,6))
ax = fig.add_axes([0,0,1,1])
ax.bar(smallest_students['District'],smallest_students['Students'])
plt.title('Top 10 Smallest Student Among All')
plt.xlabel('District')
plt.ylabel('Students')
plt.rcParams['font.size'] = '18'
plt.show()
```
