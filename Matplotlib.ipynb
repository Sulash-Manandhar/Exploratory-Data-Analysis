```python
import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np
```

```python
x = [1,2,3,4,5,6,7]
y =[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y, color = 'green', linewidth = 5, linestyle = 'dotted')
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.show()
```

```python
plt.plot(x,y,'D--r') # d for symbol(diamond), -- for line type and r for color type
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.show()
```

# MarkerSize

```python
#manual method
plt.plot(x,y,color = 'blue', marker = '*', linestyle = ':', markersize = 10) 
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.show()
```

# Alpha value

```python
#alpha value can be scaled from 0 to 1
plt.plot(x,y,'D--r', alpha = 0.4) 
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.show()
```

```python
company = ['GOOGLE', 'AMAZON', 'MSFT', 'FACEBOOK']
revenue = [90,136,89,27]
```

```python
plt.bar(company,revenue)
plt.show()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(company,revenue, label= 'Revenue')
plt.xticks(xpos,company)
plt.ylabel('Revenue(Bin)')
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

```python
profit = [40,2,34,12]
```

# Multiple bar graph

```python
plt.bar(xpos-0.2, revenue, width = 0.4, label= 'Revenue')
plt.bar(xpos+0.2, profit, width = 0.4, label= 'Profit')


plt.xticks(xpos,company)
plt.ylabel('Revenue(Bin)')
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

# Bar graph horizontal 

```python
plt.barh(xpos,revenue, label= 'Revenue')
plt.yticks(xpos,company)
plt.ylabel('Revenue(Bin)')
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

# Hist

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth =0.8) #by default number of bins is set to 10
plt.show()
```

```python
plt.hist(blood_sugar,rwidth=0.5, bins = 4)
```

```python
plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chat')

plt.hist(blood_sugar, bins =[80,100,125,150], rwidth = 0.95, color ='g')
```

# Multiple Hist 

```python
plt.xlabel('Sugar level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')

blood_sugar_man = [113, 85, 90,150, 149, 88, 93, 115, 80, 77, 82,129]
blood_sugar_women = [67, 98, 89, 120,133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_man, blood_sugar_women], bins = [80, 100, 125, 160], rwidth = 0.95, color = ['green', 'orange'], label = ['men', 'women'])
plt.legend()
plt.show()
```

# Histtype step

```python
plt.xlabel('Sugar level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')

plt.hist(blood_sugar_women, bins = [80, 100, 125, 160], rwidth = 0.95, color = 'green', label = 'women', histtype = 'step')
plt.legend()
plt.show()
```

# Horizontal hist

```python
plt.xlabel('Sugar level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')

plt.hist(blood_sugar_women, bins = [80, 100, 125, 160], rwidth = 0.95, color = 'green', label = 'women', orientation= 'horizontal')
plt.legend()
plt.show()
```

# Pie Chart

```python
exp_values = [1400,600,300,410,250]
exp_labels = ['Home Rent', 'Food', 'Phone/Telephone bills', 'Car', 'Other Utilities']
plt.pie(exp_values,labels = exp_labels)
plt.show()
```

# Explode Piechart

```python
plt.pie(exp_values, labels= exp_labels, shadow = True, autopct = '%1.1f%%', radius =1.5, explode = [0,0,0,0.1,0.2])
#autopct format for float data type
plt.show()
```

# Rotating the pie chart

```python
plt.pie(exp_values, labels= exp_labels, shadow = True, autopct = '%1.1f%%', radius =1.5, explode = [0,0,0,0.1,0.2], counterclock = True, startangle = 45)
plt.show()
```

# Saving the PieChart

```python
exp_values = [1400,600,300,410,250]
exp_labels = ['Home Rent', 'Food', 'Phone/Telephone bills', 'Car', 'Other Utilities']
plt.pie(exp_values, labels= exp_labels, shadow = True, autopct = '%1.1f%%', radius =1.5, explode = [0,0,0,0.1,0.2], counterclock = True, startangle = 45)
plt.axis('equal')
plt.savefig('piechart.jpg', bbox_inches = 'tight', pad_inches = 1, transparent = True)
#plt.savefig('piechart.pdf', bbox_inches = 'tight', pad_inches = 1, transparent = True)
plt.show()
```
