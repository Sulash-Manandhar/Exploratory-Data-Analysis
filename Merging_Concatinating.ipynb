# Join

```python
men2004 = pd.read_csv('men2004.csv')
men2008 = pd.read_csv('men2004.csv')
```

```python
men2004.head()
```

```python
men2004.shape
```

```python
men2008.head()
```

```python
men2008.shape
```

```python
len(men2004) + len(men2008) 
```

# Outer Join - union 

```python
men2004.merge(men2008, how="outer", on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
```

```python
merged_df = men2004.merge(men2008, how = "outer", on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
merged_df._merge.value_counts()
```

# Inner Join - intersection

```python
men2004.merge(men2008, how = 'inner', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
```

# Outer join without intersection - no(A) + no(B)

```python
merged_df = men2004.merge(men2008, how = 'outer', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
merged_df.head()
```

```python
merged_df.loc[merged_df._merge != 'both']
```

# Left join - n(A)

```python
men2004.merge(men2008, how = 'left', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
```

# Right join - n(B)

```python
men2004.merge(men2008, how = 'right', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
```

# Left join without intersection - no(A)

```python
merged_df = men2004.merge(men2008, how = 'outer', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
merged_df.head()
```

```python
merged_df[merged_df._merge == 'left_only']
```

# Right join without intersection - no(B)

```python
merged_df[merged_df._merge == 'right_only']
```

# Merging two df with different columns

```python
men2004 = pd.read_csv('men2004.csv')
men2008 = pd.read_csv('men2008.csv')
```

```python
men2004.columns = ['Names', 'Medals']
```

```python
men2004.head()
```

```python
men2008.head()
```

```python
merged_df = men2004.merge(men2008, how = 'outer', left_on = 'Names', right_on = 'Athlete', suffixes= ['_2004', '_2008'], indicator = True)
```

```python
merged_df
```

```python
merged_df.Names.fillna(merged_df.Athlete, inplace = True)
```

```python
merged_df
```

```python
merged_df.drop(columns = ['Athlete', '_merge'], inplace = True)
```
