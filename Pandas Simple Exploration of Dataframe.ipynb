# reading string index data

```python
summer = pd.read_csv('EDA02summer.csv',index_col="Athlete")
summer
```

```python
#bad way to select string index data
summer.iloc[4]
```

```python
#showing index
summer.index
```

```python
#best way to select data
summer.loc['NEUMANN, Paul']
```

```python
#if therre is more than one data it return dataframe not series
summer.loc['PHELPS, Michael']
```

```python
#selecting a particular row and colun
summer.loc['PHELPS, Michael','Medal']
```

```python
#selecting a row and different cols
summer.loc['PHELPS, Michael',['Medal','Event']]
```

```python
#different rows and columns
summer.loc[['PHELPS, Michael','REZAEI, Ghasem Gholamreza'],['Medal','Event']]
```

```python
summer.loc[:,['Medal','Event']]
```

```python
#loc last label is inclusive
summer.loc[:'CHASAPIS, Spiridon']
```

```python
summer.head(10)
```

```python
#shows errors if index repeat again #drawback
summer.loc[:"CHOROPHAS, Efstathios"]
```
